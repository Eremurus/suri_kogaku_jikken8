```python
%tensorflow_version 2.x
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
import time
import tensorflow as tf
from tensorflow.keras import layers, models, initializers
from tensorflow.keras.datasets import mnist
#
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
print("image size : ", train_images[0].shape)
print("image components : ", train_images[0][10])
# 0から1の浮動小数点の値に変換する
train_images = train_images.reshape(
                  (len(train_images), 28*28)).astype('float32') / 255
test_images = test_images.reshape(
                  (len(test_images), 28*28)).astype('float32') / 255
#正解ラベルの形式はワンホットベクトルに変換している
train_labels = tf.keras.utils.to_categorical(train_labels, 10)
test_labels = tf.keras.utils.to_categorical(test_labels, 10)
print("train_images : ", train_images.shape)
print("train_labels : ", train_labels.shape)
print("test_images : ", test_images.shape)
print("test_labels : ", test_labels.shape)
#
model = models.Sequential()
model.add(layers.Dense(1024, activation='relu', input_shape=(28*28,),
                       kernel_initializer=initializers.TruncatedNormal(),
                       name='relu_1'))
model.add(layers.Dense(256, activation='relu',kernel_initializer=initializers.TruncatedNormal(),
                       name='relu_2'))
model.add(layers.Dense(256, activation='relu',kernel_initializer=initializers.TruncatedNormal(),
                       name='relu_3'))
model.add(layers.Dense(10, activation='softmax', name='softmax'))
model.summary()

opt_list = ['sgd', 'adadelta', 'adagrad', 'adam', 'ftrl', 'RMSProp']
for item in opt_list:
    print(item,"の結果は以下")
    model.compile(optimizer= item, 
              loss='categorical_crossentropy', #多項分類におけるクロスエントロピー
              metrics=['acc']) #モデルの性能評価は分類の正解率ACC

    t1 = time.time()#現在時刻を取得
    history = model.fit(train_images, train_labels,
                    validation_data=(test_images, test_labels),
                    batch_size=128, #(469-1)*128 + 96 = 60000
                    epochs=10)
    print("かかった時間は",time.time() - t1)
```
